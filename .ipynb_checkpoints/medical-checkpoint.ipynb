```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('medical_examination.csv')
```

```python
df["BMI"] = df["weight"] / ((df["height"]/100) ** 2)
```

```python
df['overweight'] = (df['BMI'] > 25).astype(int)
```
